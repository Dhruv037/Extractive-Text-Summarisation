```python
from sklearn.manifold import TSNE
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
import pickle
```

```python
X_train=pd.read_excel("X_train.xlsx")
y_train=pd.read_excel("y_train.xlsx")
```

```python
model_tSNE = TSNE(n_components=2, random_state=0)
tsne_data = model_tSNE.fit_transform(X_train)
```

```python
print(tsne_data.shape)
print(y_train.shape)
tsne_data=np.concatenate((tsne_data, y_train), axis=1)
```

```python
# tsne_data = np.vstack((tsne_data.T, y_train)).T
tsne_df = pd.DataFrame(data=tsne_data, columns=("Dim_1", "Dim_2", "Label"))
sn.FacetGrid(tsne_df, hue="Label", size=30).map(plt.scatter, 'Dim_1', 'Dim_2').add_legend()
plt.show()
```

```python
filename = 'visualise_data.sav'
pickle.dump(tsne_data, open(filename, 'wb'))
```
